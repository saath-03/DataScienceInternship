```python
import pandas as pd
# import dataframe as df
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
```

```python
data = pd.read_csv('Heart.csv')
data.head()
```

```python
#Correlation matrix for the data

mat = data.corr()
print("Correlation Matrix",mat)

# Data Preprocessing

correlated_features = set()
l = len(mat.columns)
for i in range(l):
    for j in range(i):
        if abs(mat.iloc[i, j]) > 0.85:
            colname_i = mat.columns[i]
            colname_j = mat.columns[j]
            correlated_features.add(colname_i)
            correlated_features.add(colname_j)
data.drop(columns=correlated_features, inplace=True)
```

```python
#Spliting of Data

Y = data['target']
X = data.drop('target', axis=1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=30)
```

```python
#Model initialisation
Random_Forest = RandomForestClassifier(n_estimators=10, random_state=20)
Support_Vector_Machine = SVC(kernel='linear', C=1.0, random_state=20)
Logistic_Regression = LogisticRegression(random_state=20)
#Creating Ensemble Model
voting_classifier = VotingClassifier(estimators=[('rf', Random_Forest), ('svm', Support_Vector_Machine), ('lr', Logistic_Regression)],voting='hard')
```

```python
#Model Training
voting_classifier.fit(X_train, Y_train)
#Prediction
Y_pred = voting_classifier.predict(X_test)
```

```python
#Model Evaluation
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy",accuracy)
print("Confusion Matrix")
confusion_matrix(Y_test, Y_pred)
print("F1 Score",f1_score(Y_test, Y_pred))
```
